# Zad.

* Wykona kenrlową estymację gęstości 2D.
* Dobierz odpowiednią szerokość okna.

```python
import numpy as np
import matplotlib.pyplot as pl
import scipy.stats as st
from sklearn import cluster, datasets

data = np.random.multivariate_normal((0, 0), [[1, 0], [0, 1]], 100)
xmin, xmax = -3, 3
ymin, ymax = -3, 3

x, y = np.mgrid[xmin:xmax:.1, ymin:ymax:.1]
positions = np.vstack([x.ravel(), y.ravel()])
values = np.vstack([data[:, 0], data[:, 1]])
kernel = st.gaussian_kde(values)
result = np.reshape(kernel(positions).T, x.shape)

fig = pl.figure()
ax = fig.gca()
ax.set_xlim(xmin, xmax)
ax.set_ylim(ymin, ymax)

cfset = ax.contourf(x, y, result, cmap='Blues')
cset = ax.contour(x, y, result, colors='k')

ax.clabel(cset, inline=1, fontsize=10)

pl.show()
```
