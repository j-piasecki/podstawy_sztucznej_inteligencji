# Support vector regression (SVR)

```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import sklearn.linear_model as lm
from sklearn import datasets, linear_model
import matplotlib.pyplot as plt

from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR
```

Proszę narysować wykres dowolnej funkcji na odcinku $[0,3]$.

Będzie to model, który będziemy chcieli znaleźć.

```python
f = lambda x: ((x*2-1)*(x**2-2)*(x-2)+3)
x_tr = np.linspace(0, 3, 200)
y_tr = f(x_tr)
x = stats.uniform(0,3).rvs(100)
y = f(x) + stats.norm(0,0.2).rvs(len(x))
x=np.vstack(x)
x_plot = np.vstack(np.linspace(0, 10, 100))
```

Proszę wylosować kilka punktów z otoczenia funkcji $f(x)$.

Na ich podstawie będziemy wykonywać naszej presykcji.

# Wyznacz optymalne parametry

```python
svr = GridSearchCV(SVR(kernel='rbf', gamma=0.1), cv=5,
                   param_grid={"C": [0.001, 0.01, 0.1, 1, 10 ,100, 1000],
                               "gamma": [0.001, 0.01, 0.1, 1, 10 ,100, 1000]})
```

```python
svr.fit(x, y)
svr.best_params_
```

```python
# Plot outputs
plt.figure(figsize=(6,6));
axes = plt.gca()
axes.set_xlim([0,3])
axes.set_ylim([0,8])
plt.scatter(x, y,  color='black')
plt.plot(x_plot, svr.predict(x_plot), color='blue',linewidth=3)
plt.show()
```

```python
model1 = SVR(kernel='rbf', C=1, gamma=1)
model1.fit(x, y)

model2 = SVR(kernel='rbf', C=1, gamma=0.01)
model2.fit(x, y)

model3 = SVR(kernel='rbf', C=1, gamma=10)
model3.fit(x, y)

model4 = SVR(kernel='rbf', C=0.1, gamma=1)
model4.fit(x, y)

model5 = SVR(kernel='rbf', C=10, gamma=1)
model5.fit(x, y)
```

```python
# Plot outputs
plt.figure(figsize=(6,6));
axes = plt.gca()
axes.set_xlim([0,3])
axes.set_ylim([0,8])
plt.scatter(x, y,  color='black')
plt.plot(x_plot, model1.predict(x_plot), color='blue', linewidth=3)
plt.plot(x_plot, model2.predict(x_plot), color='red', linewidth=3)
plt.plot(x_plot, model3.predict(x_plot), color='green', linewidth=3)
plt.plot(x_plot, model4.predict(x_plot), color='cyan', linewidth=3)
plt.plot(x_plot, model5.predict(x_plot), color='magenta', linewidth=3)
plt.show()
```

# Zad

Wyznacz optymalne parametry dla modelu SVR.

```python
import pandas as pd
df_adv = pd.read_csv('Advertising.csv', index_col=0)
X = df_adv[['TV', 'radio','newspaper']]
y = df_adv['sales']
df_adv.head()
```

```python
from sklearn import  metrics

svr = GridSearchCV(SVR(kernel='rbf', gamma=0.1), cv=5,
                   param_grid={"C": [0.01, 0.1, 1, 10, 100, 1000, 10000],
                               "gamma": [0.000001, 0.00001, 0.0001, 0.001, 0.01, 0.1]})

svr.fit(X, y)
print("R^2: {}".format(metrics.r2_score(y, svr.predict(X))))
svr.best_params_
```
