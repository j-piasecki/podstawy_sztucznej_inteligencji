# Wielowymiarowy rozkład normalny

```python
import matplotlib.pylab as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.mlab as mlab
from matplotlib import cm
```

# Zad. 
Narysujmy gęstość dwuwymiarowego rozkładu Normalnego o parametach 

* n=[0,0], cov = [[1,0],[0,1]]
* n=[0,0], cov = [[2,0],[0,1]]
* n=[0,0], cov = [[1,0],[0,2]]

```python
from scipy.stats import multivariate_normal

x = np.linspace(-5, 5, 500)
y = np.linspace(-5, 5, 500)
X, Y = np.meshgrid(x, y)

pos = np.empty(X.shape + (2,))
pos[:, :, 0] = X
pos[:, :, 1] = Y

#rv = multivariate_normal([0, 0], [[1, 0], [0, 1]])
#rv = multivariate_normal([0, 0], [[2, 0], [0, 1]])
rv = multivariate_normal([0, 0], [[1, 0], [0, 2]])
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, rv.pdf(pos),cmap='viridis',linewidth=0)

plt.show()
```

# Zad 
Wylosujmy próbkę z dwuwymiarowego rozkładu Normalnego o parametach 

* n=[0,0], cov = [[1,0],[0,1]]
* n=[0,0], cov = [[2,0],[0,1]]
* n=[0,0], cov = [[1,0],[0,2]]

i narysuj odpowiedni wykres

```python
from scipy.stats import multivariate_normal

ax = plt.axes()

ax.set_xlim([-5, 5])
ax.set_ylim([-5, 5])

x = np.linspace(-5, 5, 500)
y = np.linspace(-5, 5, 500)
X, Y = np.meshgrid(x, y)

pos = np.empty(X.shape + (2,))
pos[:, :, 0] = X
pos[:, :, 1] = Y

rv = multivariate_normal([0, 0], [[1, 0], [0, 1]])
#rv = multivariate_normal([0, 0], [[2, 0], [0, 1]])
#rv = multivariate_normal([0, 0], [[1, 0], [0, 2]])
S = rv.rvs(500)
plt.scatter(S[:, 0], S[:, 1])

plt.show()
```
